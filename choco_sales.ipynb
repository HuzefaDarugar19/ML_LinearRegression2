```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

df= pd.read_csv('flavours.csv')
r1=df.columns 
#(['Company \n(Maker-if known)', 'Specific Bean Origin\nor Bar Name',
     #  'REF', 'Review\nDate', 'Cocoa\nPercent', 'Company\nLocation', 'Rating',
       #'Bean\nType', 'Broad Bean\nOrigin'],''' 
df.head(50)
```

***We will where are the cocoa beans coming from and see which country is the  most productive for manufacturing***

```python
#'Broad Bean\nOrigin'
#'Cocoa\nPercent'
#'Bean\nType'
df[['Broad Bean\nOrigin','Cocoa\nPercent']]
df['Cocoa\nPercent'] = pd.to_numeric(df['Cocoa\nPercent'], errors='coerce')

average_production = df.groupby('Broad Bean\nOrigin')['Cocoa\nPercent'].mean()
print(average_production)

```

***lets visualize this data***

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Reset the index for Seaborn compatibility
average_production = average_production.reset_index()

# Create a bar plot
sns.barplot(data=average_production, x='Broad Bean\nOrigin', y='Cocoa\nPercent', palette='coolwarm')
plt.title('Average Cocoa Percentage by Broad Bean Origin', fontsize=16)
plt.xlabel('Broad Bean Origin', fontsize=12)
plt.ylabel('Average Cocoa Percentage', fontsize=12)
plt.xticks(rotation=45)
plt.show()

```
