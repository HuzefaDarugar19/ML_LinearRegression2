```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

X=np.array([1,2,3,4,5])
Y=np.array([1,4,9,16,25])

m=0
b=0

iteration=10000
rate=0.01
n=len(X)
# y= m*x+b, which is equation of line 
for i in range(iteration):
   y_pred=m*X+b
   error=y_pred-Y
   cost =(1/n)*sum(error**2)
   m_gradient=(2/n)*sum(error*X)
   b_gradient=(2/n)*sum(error)
 
#updating parameters 
   m-=m_gradient*rate
   b-=b_gradient*rate

   if(i%200==0):
     print(f"{cost:.4f},{m:.4f},{b:.4f}")
     print(f"y={m*X+b}")
print("hello")

#sns.lineplot(x=X,y=Y , color='red')
#sns.scatterplot(x=X,y=Y,color='blue')
sns.lineplot(x=X, y=m*X + b, color='red', label='Final Model')
sns.scatterplot(x=X, y=m*X + b, color='blue', label='points')
```
